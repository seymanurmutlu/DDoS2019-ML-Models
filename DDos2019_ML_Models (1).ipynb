```python
#handling categorical data
df = df.rename(columns={" Label": "Label"})
df.info()
```

```python
df.Label.unique()
```

```python
df['Label'] = df['Label'].replace('BENIGN', '0')
df['Label'] = df['Label'].replace('WebDDoS', '1')
df['Label'] = df['Label'].replace('DrDoS_UDP', '2')
df['Label'] = df['Label'].replace('Syn', '3')
df['Label'] = df['Label'].replace('DrDoS_MSSQL', '4')
df['Label'] = df['Label'].replace('DrDoS_LDAP', '5')
df['Label'] = df['Label'].replace('DrDoS_SNMP', '6')
df['Label'] = df['Label'].replace('UDP-lag', '7')
df['Label'] = df['Label'].replace('DrDoS_DNS', '8')
df['Label'] = df['Label'].replace('DrDoS_NTP', '9')
df['Label'] = df['Label'].replace('DrDoS_NetBIOS', '10')
df['Label'] = df['Label'].replace('DrDoS_SSDP', '11')
df['Label'] = df['Label'].replace('TFTP', '12')
df['Label'] = df['Label'].astype('int')
```
